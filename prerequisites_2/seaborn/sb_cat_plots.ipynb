```python
import seaborn as sns
%matplotlib inline

s = sns.load_dataset('tips')
s.head(5)
```

```python
#barplot
# plots categorical data based on some function

sns.barplot(x='sex',y='total_bill',data=s)
```

```python
#countplot

sns.countplot(x='smoker',data=s) # counts the number of occurances 
```

```python
#boxplot
sns.boxplot(x="day", y="total_bill", data=s)
```

```python
# with hue
sns.boxplot(x="day", y="total_bill", data=s,hue='smoker')
```

```python
#violin plot shows the distribution of quantitative data acress several plots
#violin plots featurea a kde of distribution rather than actual data points in barplot 
sns.violinplot(x="day", y="total_bill", data=s,palette='rainbow')
```

```python
# strip plot
sns.stripplot(x="day", y="total_bill", data=s)
```

```python
# swarm plot - > combination of strip and violin plot 
sns.swarmplot(x="day", y="total_bill", data=s)
```

```python
sns.catplot(x='sex',y='total_bill',data=s,kind='bar') #factorplot now named as catplot
```
