```python
import seaborn as sns
%matplotlib inline

t= sns.load_dataset('tips')
t.head(5)
```

```python
sns.lmplot(x='total_bill',y='tip',data=t) # lmplot is a type of regression plot ...allows to display linear models
```

```python
sns.lmplot(x='total_bill',y='tip',data=t,col='sex') # extra parameters can be added 
```
