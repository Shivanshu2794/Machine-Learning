```python
import seaborn as sns
%matplotlib inline

t = sns.load_dataset('tips')
f = sns.load_dataset('flights')
f.head(5)
```

```python
t.head(5)
```

```python
# heatmap requires data to be in matrix form
t.corr() #matrix form
```

```python
sns.heatmap(t.corr(),annot=True) #place actual values as well
```

```python
pv=f.pivot_table(values='passengers',index='month',columns='year') # flight data in matrix form
pv
```

```python
sns.heatmap(pv)
```

```python
# cluster map 
# produces heirarichal clustering for a heatmapped data
sns.clustermap(pv)
```
