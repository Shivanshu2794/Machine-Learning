```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

i = sns.load_dataset('iris')

t=sns.load_dataset('tips')
i.head(5)
```

```python
# Grid -> allow yto map plots to rows and columns of a grid
#pairgrid-> subplot grid for plotting pair wise relationship 

grid = sns.PairGrid(i) # to create pair grids accroding to data sets 
grid.map(plt.scatter) # mapping to plot data in grids
```

```python
# mapping of pairgrids can be changed 
# example : all diagonal plotted as histogram and rest are plotted as scatter plot
grid = sns.PairGrid(i)
grid.map_diag(plt.hist)
grid.map_upper(plt.scatter)
grid.map_lower(plt.scatter)

# pairplot belongs to pair grid 
```

```python
grid = sns.FacetGrid(t, col="time",  row="smoker")
grid.map(plt.hist, "total_bill")
```

```python
grid = sns.JointGrid(x="total_bill", y="tip", data=t) #joint grid is a simple version of joint plot
grid.plot(sns.regplot, sns.distplot)
```
